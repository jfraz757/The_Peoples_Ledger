#### The Spreadsheet from the KY Finance and Revenue Cabinet will need to converted from .xlsx to .csv

```python
# Convert KY_Finance_&_Admin data from .xcel to .csv 
# Refined code was edited using Claude.ai

from pathlib import Path
import pandas as pd
from datetime import datetime

def convert_ky_fa_xlsx_to_csv():
    folder = Path('Spreadsheets/KY_Finance_&_Administration')
    
    # Find the .xlsx file in the folder
    xlsx_files = list(folder.glob('*.xlsx'))
    
    if not xlsx_files:
        print("No .xlsx file found in KY_Finance_&_Administration.")
        return
    if len(xlsx_files) > 1:
        print(f"Multiple .xlsx files found: {[f.name for f in xlsx_files]}. Using the first one.")
    
    input_path = xlsx_files[0]
    
    # Build output filename with current month and year (when function is run)
    now = datetime.now()
    month = now.strftime('%B')   # e.g. "May"
    year = now.strftime('%Y')    # e.g. "2026"
    
    output_path = folder / f"KY_F&A_{month}_{year}.csv"
    
    # Convert
    df = pd.read_excel(input_path)
    df.to_csv(output_path, index=False)
    print(f"Created '{output_path.name}'")
    
    # Delete the original .xlsx
    input_path.unlink()
    print(f"Deleted '{input_path.name}'")

convert_ky_fa_xlsx_to_csv()
```

#### This Function will rename the .csv file in the Louiville_HRC folder and KY_Transporation_Cabinet folder to a filename with the month and year the file was ran.

```python
#code was refinied using Claude.ai

from pathlib import Path
from datetime import datetime

def rename_csv_files():
    now = datetime.now()
    month = now.strftime('%B')  # e.g. "May"
    year = now.strftime('%Y')   # e.g. "2026"

    folders_and_names = {
        'Spreadsheets/Louisville_HRC': f'Louisville_HRC_{month}_{year}.csv',
        'Spreadsheets/KY_Transportation_Cabinet': f'KY_Trans_Cab_{month}_{year}.csv'
    }

    for folder_path, new_name in folders_and_names.items():
        folder = Path(folder_path)

        csv_files = list(folder.glob('*.csv'))

        if not csv_files:
            print(f"No .csv file found in '{folder.name}'.")
            continue
        if len(csv_files) > 1:
            print(f"Multiple .csv files found in '{folder.name}': {[f.name for f in csv_files]}. Using the first one.")

        old_path = csv_files[0]
        new_path = folder / new_name

        old_path.rename(new_path)
        print(f"'{folder.name}': '{old_path.name}' -> '{new_name}'")

rename_csv_files()
```
